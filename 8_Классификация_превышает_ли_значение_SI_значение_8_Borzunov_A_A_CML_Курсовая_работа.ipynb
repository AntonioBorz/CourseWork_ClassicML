# **8. Классификация - превышает ли значение SI значение 8**

```python
# Импортируем необходимые библиотеки
from google.colab import drive
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, cross_val_score, RandomizedSearchCV
from sklearn.preprocessing import RobustScaler
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import (accuracy_score, precision_score, recall_score,
                            f1_score, roc_auc_score, confusion_matrix,
                            classification_report, roc_curve, auc)
import xgboost as xgb
!pip install catboost
from catboost import CatBoostClassifier
import joblib
from scipy import stats

import warnings
warnings.filterwarnings('ignore')
```

1. Загрузка данных после EDA

```python
# Загружаем данные
drive.mount('/content/drive')
df = pd.read_excel('/content/drive/MyDrive/EDAprocessed_data.xlsx')
```

```python
# Проверим датасет
print(f"Количество строк: {len(df)}")
print(f"Количество столбцов: {len(df.columns)}")

# Посмотрим первые строчки
df.head()
```

2. Подготовка данных

```python
# Устанавливаем пороговое значение (SI > 8)
threshold_si = 8.0
df['SI_above_8'] = (df['SI'] > threshold_si).astype(int)
target_class_si_8 = 'SI_above_8'

print(f"Пороговое значение SI: {threshold_si}")
print(f"Распределение классов:\n{df[target_class_si_8].value_counts(normalize=True)}")
```

```python
# Используем те же признаки, что и для регрессии (исключая целевые переменные)
exclude_cols = ['IC50, mM', 'log_IC50, mM', 'CC50, mM', 'log_CC50, mM', 'SI', 'log_SI']
X_si_8 = df.drop(exclude_cols + [target_class_si_8], axis=1)
y_si_8 = df[target_class_si_8]
```

```python
# Разделяем данные стратифицированно
X_train_si_8, X_test_si_8, y_train_si_8, y_test_si_8 = train_test_split(
    X_si_8, y_si_8, test_size=0.2, random_state=42, stratify=y_si_8
)
print(f"Обучающая выборка: {X_train_si_8.shape}, Тестовая: {X_test_si_8.shape}")
```

```python
# Масштабируем данные
scaler_si_8 = RobustScaler()
X_train_si_8_scaled = scaler_si_8.fit_transform(X_train_si_8)
X_test_si_8_scaled = scaler_si_8.transform(X_test_si_8)
```

3. Определение и подготовка набора моделей

```python
# Определяем модели классификации с учетом несбалансированности
models_si_8 = {
    'RandomForest': RandomForestClassifier(
        n_estimators=300, random_state=42, n_jobs=-1, class_weight='balanced'
    ),
    'GradientBoosting': GradientBoostingClassifier(n_estimators=200, random_state=42),
    'XGBoost': xgb.XGBClassifier(
        eval_metric='logloss', random_state=42,
        scale_pos_weight=sum(y_train_si_8 == 0) / sum(y_train_si_8 == 1)
    ),
    'CatBoost': CatBoostClassifier(
        iterations=300, verbose=False, random_state=42,
        auto_class_weights='Balanced'
    ),
    'LogisticRegression': LogisticRegression(
        max_iter=5000, random_state=42, class_weight='balanced', n_jobs=-1
    ),
    'KNN': KNeighborsClassifier(n_jobs=-1)
}
```

4. Оценка моделей и выбор лучшей модели

```python
# Создаем функцию для оценки классификаторов
def evaluate_classifiers(models, X_train, X_test, y_train, y_test, cv_folds=5):
    """Оцениваем модели классификации с помощью кросс-валидации и тестовой выборки"""
    results = {}

    for name, model in models.items():
        print(f"\nОцениваем модель: {name}")

        try:
            # Кросс-валидация (ROC-AUC)
            cv_scores = cross_val_score(model, X_train, y_train, cv=cv_folds,
                                       scoring='roc_auc', n_jobs=-1)

            # Обучение модели
            model.fit(X_train, y_train)

            # Предсказания
            y_pred_train = model.predict(X_train)
            y_pred_test = model.predict(X_test)
            y_proba_test = model.predict_proba(X_test)[:, 1]

            # Вычисление метрик
            results[name] = {
                'CV_AUC_mean': cv_scores.mean(),
                'CV_AUC_std': cv_scores.std(),
                'Accuracy': accuracy_score(y_test, y_pred_test),
                'Precision': precision_score(y_test, y_pred_test),
                'Recall': recall_score(y_test, y_pred_test),
                'F1': f1_score(y_test, y_pred_test),
                'ROC_AUC': roc_auc_score(y_test, y_proba_test),
                'Model': model
            }

            print(f"Test ROC-AUC: {results[name]['ROC_AUC']:.4f}")
            print(f"Test F1: {results[name]['F1']:.4f}")

        except Exception as e:
            print(f"Ошибка при оценке модели: {str(e)}")

    return results
```

```python
# Для моделей, не требующих масштабирования (деревья и проч.)
tree_models_si_8 = {k: v for k, v in models_si_8.items() if k not in ['LogisticRegression', 'KNN']}
results_original_si_8 = evaluate_classifiers(tree_models_si_8, X_train_si_8, X_test_si_8, y_train_si_8, y_test_si_8)

# Для моделей, требующих масштабирования
scaled_models_si_8 = {k: v for k, v in models_si_8.items() if k in ['LogisticRegression', 'KNN']}
results_scaled_si_8 = evaluate_classifiers(scaled_models_si_8, X_train_si_8_scaled, X_test_si_8_scaled,
                                         y_train_si_8, y_test_si_8)

# Объединяем результаты
all_results_si_8 = {**results_original_si_8, **{f"{k}_scaled": v for k, v in results_scaled_si_8.items()}}
results_df_si_8 = pd.DataFrame(all_results_si_8).T
results_df_si_8 = results_df_si_8.sort_values('ROC_AUC', ascending=False)
```

```python
# Выбираем лучшую модель
best_model_name_si_8 = results_df_si_8.index[0]
best_model_si_8 = results_df_si_8.loc[best_model_name_si_8, 'Model']
best_model_type = type(best_model_si_8).__name__

print(f"Лучшая модель: {best_model_name_si_8} ({best_model_type})")
print(f"Test ROC-AUC: {results_df_si_8.loc[best_model_name_si_8, 'ROC_AUC']:.4f}")
print(f"Test F1: {results_df_si_8.loc[best_model_name_si_8, 'F1']:.4f}")
```

```python
# Донастраиваем модель
if 'CatBoost' in best_model_name_si_8:
    param_grid = {
        'iterations': [500, 700],
        'depth': [6, 8],
        'learning_rate': [0.05, 0.1],
        'l2_leaf_reg': [3, 5]
    }
elif 'XGBoost' in best_model_name_si_8:
    param_grid = {
        'n_estimators': [300, 400],
        'max_depth': [5, 7],
        'learning_rate': [0.05, 0.1],
        'subsample': [0.8, 0.9]
    }
elif 'RandomForest' in best_model_name_si_8:
    param_grid = {
        'n_estimators': [300, 400],
        'max_depth': [10, 20, None],
        'min_samples_split': [2, 5],
        'min_samples_leaf': [1, 2],
        'max_features': ['sqrt', 'log2']
    }
elif 'GradientBoosting' in best_model_name_si_8:
    param_grid = {
        'n_estimators': [200, 300],
        'learning_rate': [0.1, 0.2],
        'max_depth': [3, 5],
        'subsample': [0.8, 0.9]
    }
else:
    param_grid = {}

if param_grid:
    random_search = RandomizedSearchCV(
        estimator=best_model_si_8,
        param_distributions=param_grid,
        n_iter=15,
        cv=3,
        scoring='roc_auc',
        n_jobs=-1,
        random_state=42,
        verbose=1
    )

    # Выбираем правильные данные
    if 'scaled' in best_model_name_si_8:
        X_train_final, X_test_final = X_train_si_8_scaled, X_test_si_8_scaled
    else:
        X_train_final, X_test_final = X_train_si_8, X_test_si_8

    random_search.fit(X_train_final, y_train_si_8)
    tuned_model = random_search.best_estimator_

    # Оцениваем настроенную модель
    y_proba_tuned = tuned_model.predict_proba(X_test_final)[:, 1]
    tuned_auc = roc_auc_score(y_test_si_8, y_proba_tuned)
    y_pred_tuned = tuned_model.predict(X_test_final)
    tuned_f1 = f1_score(y_test_si_8, y_pred_tuned)

    original_auc = results_df_si_8.loc[best_model_name_si_8, 'ROC_AUC']
    original_f1 = results_df_si_8.loc[best_model_name_si_8, 'F1']

    print(f"Лучшие параметры: {random_search.best_params_}")
    print(f"Test ROC-AUC после настройки: {tuned_auc:.4f} (было {original_auc:.4f})")
    print(f"Test F1 после настройки: {tuned_f1:.4f} (было {original_f1:.4f})")

    # Выбираем лучшую модель на основе ROC-AUC
    if tuned_auc > original_auc:
        final_model = tuned_model
        print("Настроенная модель лучше исходной по ROC-AUC.")
    else:
        final_model = best_model_si_8
        print("Исходная модель лучше или равна настроенной по ROC-AUC.")
else:
    final_model = best_model_si_8

y_pred_final = final_model.predict(X_test_final)
y_proba_final = final_model.predict_proba(X_test_final)[:, 1]
```

5. Анализ результатов и оценка важности признаков

```python
# Важность признаков
if hasattr(final_model, 'feature_importances_'):
    importances = final_model.feature_importances_
elif hasattr(final_model, 'coef_'):
    importances = np.abs(final_model.coef_[0])
else:
    # Для моделей без атрибута важности
    from sklearn.inspection import permutation_importance
    result = permutation_importance(final_model, X_test_final, y_test_si_8, n_repeats=10, random_state=42)
    importances = result.importances_mean

# Нормализация важности
importances = 100.0 * (importances / importances.max())
feature_importance_si_8 = pd.DataFrame({
    'Feature': X_si_8.columns,
    'Importance': importances
}).sort_values('Importance', ascending=False)

print("Топ-10 важных признаков для классификации:")
print(feature_importance_si_8.head(10))
```

```python
# Сравнение моделей по ROC-AUC
plt.figure(figsize=(14, 8))
top_models = results_df_si_8.head(8)
model_names = [name.replace('_scaled', '*') for name in top_models.index]
roc_auc_scores = top_models['ROC_AUC'].values
colors = plt.cm.viridis(np.linspace(0.2, 0.8, len(model_names)))
bars = plt.bar(model_names, roc_auc_scores, color=colors, edgecolor='black', alpha=0.85)
plt.xticks(fontsize=10)
plt.ylabel('Test ROC-AUC', fontsize=12)
plt.title('Сравнение моделей по ROC-AUC для классификации SI (8)', fontsize=14, pad=20)
plt.grid(True, alpha=0.3, linestyle='--')
plt.ylim(min(roc_auc_scores) - 0.05, max(roc_auc_scores) + 0.05)

# Добавляем значения на столбцы
for bar, score in zip(bars, roc_auc_scores):
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height-0.02, f'{score:.3f}',
             ha='center', va='top', color='white', fontsize=10, fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
# Сравнение моделей по F1-Score
plt.figure(figsize=(14, 8))
f1_scores = top_models['F1'].values
bars = plt.bar(model_names, f1_scores, color=colors, edgecolor='black', alpha=0.85)
plt.xticks(fontsize=10)
plt.ylabel('Test F1-Score', fontsize=12)
plt.title('Сравнение моделей по F1-Score для классификации SI (8)', fontsize=14, pad=20)
plt.grid(True, alpha=0.3, linestyle='--')
plt.ylim(min(f1_scores) - 0.05, max(f1_scores) + 0.05)

# Добавляем значения на столбцы
for bar, score in zip(bars, f1_scores):
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height-0.02, f'{score:.3f}',
             ha='center', va='top', color='white', fontsize=10, fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
# Матрица ошибок
plt.figure(figsize=(8, 6))
cm = confusion_matrix(y_test_si_8, y_pred_final)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['SI ≤ 8', 'SI > 8'],
            yticklabels=['SI ≤ 8', 'SI > 8'])
plt.xlabel('Предсказанные значения')
plt.ylabel('Реальные значения')
plt.title('Матрица ошибок')
plt.show()
```

```python
# ROC-кривая
plt.figure(figsize=(10, 8))
y_proba_final = final_model.predict_proba(X_test_final)[:, 1]
fpr, tpr, thresholds = roc_curve(y_test_si_8, y_proba_final)
roc_auc = auc(fpr, tpr)
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC кривая (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC-кривая')
plt.legend(loc="lower right")
plt.grid(True, alpha=0.3)
plt.show()
```
